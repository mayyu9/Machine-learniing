## MatplotLib

Matplotlib is a plotting library for the python programming language and its numerical mathematics extension Numpy.
it provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt or GTK+

some of the ajor plots of Matplotlib are
# generally easy to get started for simple plots
# great control of every element ina figure
# Support for custom label and texts
# High-Quality output in many formats
# very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
# simple examples

x = np.arange(0,10,1)
y = np.arange(11,21,1)
```

```python
# plot the graph using scatter
## plt scatter - this use for drawing the graph

plt.scatter(x,y, c='b')
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
# plt.savefig('graphin2D.jpeg')
```

```python
y=x*x
```

```python
# plt plot
plt.plot(x,y, 'r*', linestyle='dashed', markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
# Creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'g*')

plt.subplot(2,2,3)
plt.plot(x,y, 'ro')

plt.subplot(2,2,4)
plt.plot(x,y,'b*--')
```

```python
x = np.arange(1,11)
y = 3*x+5
plt.title('Multiplication demo')
plt.xlabel('X axis caption')
plt.ylabel('Y axis caption')
plt.plot(x,y)
```

```python
# compute the x and y coordinates for points on a Sin curve
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title('Sine wave')
plt.plot(x,y)
```

```python
# Subplot()

# compute the x and y coordinates for points on sine and cosine waves

x = np.arange(0,5*np.pi, 0.1)
y_sine = np.sin(x)
y_cosine = np.cos(x)

# set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active

plt.subplot(2,1,1)

# make the first plot
plt.plot(x,y_sine, 'r--')
plt.title('sine wave')

# set the second subplot as active and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cosine, 'g--')
plt.title('cosine wave')

#show the figure
plt.show()
```

```python
# Bar plot
# x and y should be of same size, mismatch of size is not allowed

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2, y2, color='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Bar graph')

plt.show()
```

## Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

## Box plot using Matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data, vert=True, patch_artist= True)
```

## Pie Chart

```python
# Data to plot

labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 310, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0) # explode 1st slice

# plot
plt.pie(sizes, explode =explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
