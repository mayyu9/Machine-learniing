## Seaborn Tutorial

## Distribution plots

Distribution plots will allow us to plot and analysis the varient features.

- distplot -> this is useful when we have more than 2 or 3 features.
- joinplot -> this is useful in analyzing the bi varinet analysis.
- pairplot -> this is useful when we have more than 2 or 3 features.

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

## Correlation with Heatmap

A correlation heatmap use colored cells, typically in a monochromatic scale, to show a 2D correlation matric(table) between two discrete dimensions or event types. It is very important in Feature Selection

- correlation helps us to find out the inter relation between all the features 
- correlation can be applied only when values are  floating point and integer types are there

- correaltion doesn't work when the values are categorical features, because categorical features are object type

- correlation value will always ranging from -1 to +1, this is specifially called as coefficient correlation



```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Joint Plot

A joint plot allows us to study the relationshipbetween 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot.

## Univarient Analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg') # regressison
```

## Pair plot

Features are more than 2, then we use pair plot
A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this pairs plots are just elaborations on this, showing all the variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

## Dist plots

Dist plots helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

## Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we discuss about

- boxplot
- violinplot
- countplot
- bar plot

```python
## Count plot
```

```python
sns.countplot('sex', data=df)
```

```python
sns.countplot(y='smoker', data=df)
```

```python
# Bar plot

sns.barplot(y='total_bill', x='sex', data= df)
```

## Box plot

A box and whisker plot (some times called a box plot) is a graph that presents information from a five-number summary.

```python
sns.boxplot('smoker', 'total_bill', data=df)
```

```python

sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python

# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

## Violin plot

Violin plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot


```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
```
